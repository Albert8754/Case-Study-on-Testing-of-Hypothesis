# The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.

```python
x=np.array(data['Sales_After_digital_add(in $)'])
```

```python
y=np.array(data['Sales_before_digital_add(in $)'])
```

we made an array of x and y to check ttest.we make null hypothesis as no increase in sales and alternative as increase in sales

```python
stats.ttest_ind(x,y,alternative='greater')
```

pvalue=1.3071840034523225e-16 Hence we reject our null hypothesis. and there is increase in sales after stepping into digital marketing

# The company needs to check whether there is any dependency between the features “Region” and “Manager”.

Here null hypothesis:  there is no dependency between the features “Region” and “Manager”.
Alternative Hypothesis:  there is dependency between the features “Region” and “Manager”.

```python
data['Region'].value_counts()
```

```python
data['Manager'].value_counts()
```

In order to compute the Chi-square test statistic, we would need to construct a contingency table.

```python
pd.crosstab(data.Region, data.Manager)
```

The numbers in this table represent frequencies.

```python
ct = pd.crosstab(data.Region, data.Manager)
stat,p,dof,expected=stats.chi2_contingency(ct)
print(f'chisquare value:{stat}')
print(f'p value:{p}')
print(f'degree of freedom:{dof}')
print(f'expected values:{expected}')
```

we got p value as .549 which is greater than .05 so weaccept null hypothesis.therefore both are independent
